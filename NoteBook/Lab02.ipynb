```python
#Crear un dataframe con 10000000 de registro
cantidad = 1000000
```

```python
import random
```

```python
influencers = ["@AnaCancun", "@LuisPlayas", "@MarthaTravel", "@CarolinaSol", "@SofiaRiviera", "@PedroMar", "@ElenaCaribe", "@DiegoAventura"]
tipos = ["foto", "reel", "carrusel", "historia"]
```

```python
# Creación del DataFrame
dfMillon = pd.DataFrame({
    "influencer" : [random.choice(influencers) for _ in range(cantidad)],
    "likes" : [random.randint(50, 50000) for _ in range(cantidad)],
    "comentarios" : [random.randint(0, 2000) for _ in range(cantidad)],
    "fecha" : [f"2026-0{random.randint(1,5)}-{random.randint(10,28)}" for _ in range(cantidad)],
    "tipo_publicacion" : [random.choice(tipos) for _ in range(cantidad)]
})
```
